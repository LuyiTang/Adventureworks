## 3) Analyse et Vizualisation

```python
df['text'] = df['name'] + '<br>order amount ' + (df['OrderAmount']).astype(str)
limits = [(1,3),(4,8),(9,24),(25,50),(51,264)]
colors = ["black","royalblue","crimson","lightseagreen","orange"]
cities = df["City"]
scale = 1

fig = go.Figure()

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[(df["OrderAmount"]>=lim[0]) & (df["OrderAmount"]<=lim[1])]
    fig.add_trace(go.Scattergeo(
        locationmode = 'USA-states',
        lon = df_sub['lon'],
        lat = df_sub['lat'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['OrderAmount'],
            color = colors[i],
            line_color='rgb(40,40,40)',
            line_width=0.5,
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1])))

fig.update_layout(
        title_text = 'Total order number by city ',
        showlegend = True,
        geo = dict(
            scope = 'usa',
            landcolor = 'rgb(217, 217, 217)',
        )
    )

fig.show()
```
